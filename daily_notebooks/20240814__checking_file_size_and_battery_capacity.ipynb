```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import colors
```

```python
AUDIOMOTH_BIT_DEPTH = 16
BITS_PER_BYTE = 8
```

```python
def get_file_size(sampling_rate, recording_duration_secs):
    bit_rate = sampling_rate * AUDIOMOTH_BIT_DEPTH 
    byte_rate = bit_rate / BITS_PER_BYTE

    file_size = byte_rate * recording_duration_secs
    return file_size

def get_daily_data_rate(file_size, cycle_length_of_recording_minutes, recording_num_hours_per_day):
    num_cycles_in_hour = 60/cycle_length_of_recording_minutes
    return file_size*num_cycles_in_hour*recording_num_hours_per_day
```

```python
sampling_rate = 192000
recording_duration_secs = 1795

file_size = get_file_size(sampling_rate, recording_duration_secs)
print(f'Audiomoth file size: {file_size/1e6}MB')
cycle_length = 30
recording_num_hours_per_day = 24

data_rate = get_daily_data_rate(file_size, cycle_length, recording_num_hours_per_day)
print(f'Daily data intake: {data_rate/1e9}GB')
sd_card_size = np.arange(0, 640, 128)

plt.figure(figsize=(6,4))
plt.title(f'Continuous 192kHz SD card life')
plt.plot(sd_card_size, sd_card_size/(data_rate/1e9), marker='.')
plt.ylabel('Number of days')
plt.xlabel('SD card capacity')
plt.grid(which='both')
plt.show()
```

```python
daily_energy_consumption = 580
battery_capacity = 2450

working_battery_capacity = 3*np.arange(0, 1.1, 0.1)*battery_capacity

plt.figure(figsize=(6,4))
plt.title(f'Continuous 192kHz battery life ({daily_energy_consumption}mAh consumed / day)')
plt.plot(working_battery_capacity, working_battery_capacity/daily_energy_consumption, marker='.')
plt.ylabel('Number of days')
plt.xlabel('Overall battery capacity')
plt.grid(which='both')
plt.show()
```
